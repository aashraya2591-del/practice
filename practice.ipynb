```python
# Task 1: Check if all numbers in a list are unique
def all_unique(numbers):
    return len(numbers) == len(set(numbers))

# Test the function
print(all_unique([1, 2, 3, 4]))  # True
print(all_unique([1, 2, 2, 3]))  # False
```
